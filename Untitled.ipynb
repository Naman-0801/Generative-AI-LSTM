```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
import re
```

```python
with open("shakespeare.txt", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
text = text.lower()
text = re.sub(r"[^a-z\s]", "", text)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

total_words = len(tokenizer.word_index) + 1
```

```python
input_sequences = []
tokens = tokenizer.texts_to_sequences([text])[0]

sequence_length = 20
```

```python
for i in range(sequence_length, len(tokens)):
    seq = tokens[i-sequence_length:i+1]
    input_sequences.append(seq)

input_sequences = np.array(input_sequences)
```

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]
```

```python
model = Sequential([
    Embedding(total_words, 100, input_length=sequence_length),
    LSTM(150),
    Dense(total_words, activation='softmax')
])

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)


model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=3)

X = X[:200000]
y = y[:200000]

model.fit(
    X, y,
    epochs=15,
    batch_size=128,
    callbacks=[early_stop]
)
```

```python
def generate_text(seed_text, next_words=30):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences(
            [token_list],
            maxlen=sequence_length,
            padding='pre'
        )

        predicted = np.argmax(model.predict(token_list, verbose=0))
        output_word = tokenizer.index_word.get(predicted, "")

        seed_text += " " + output_word

    return seed_text
```

```python
print(generate_text("to be or not to be", 40))
print(generate_text("love is a smoke", 40))
```
